```python
#l is list of marks of student#
#c is candy for eAach student#
#logic is if a student have more marks then his neigh then he get +1 candy then him#
l=eval(input("Enter: "))
c=[1]*len(l)
for i in range(len(l)):
    if i==0:
        c[i]=1
    elif l[i]>l[i-1]:
        c[i]=c[i-1]+1
print(c)
for i in range(len(l),-1,-1):
    if i==len(l) or i==len(l)-1:
        pass
    elif l[i]>l[i+1]:
        a=c[i+1]+1
        if c[i]>a:
            pass
        else:
            c[i]=a
print(c)
totalcandy=sum(c)
print(totalcandy)
```
