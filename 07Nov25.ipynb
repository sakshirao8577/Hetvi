```python
#water in units of hight
h=eval(input("Enter: "))
l=[0]*len(h)
r=[0]*len(h)
sum=0
for i in range (-2,-len(h)-1,-1):
    m=[]*2
    m.insert(0,h[i])
    m.insert(1,r[i+1]
    r[i]=max(m)
    m.clear()
l[0]=h[0]
for i in range (1,len(h)):
    m=[]*2
    m.append(h[i])
    m[1]=l[i-1]
    l[i]=max(m)
    m.clear()
for i in range(len(h)):
    m=[]*2
    m[0]=r[i]
    m[1]=l[i]
    sum+=min(m)-h[i]
    m.clear()
r[-1]=h[-1]
print(r)
print(l)
print(sum)
```
