```python
d={'A':['aaaaa','a','aa','aaa'],'B':['bb','bbbbbbbb','bbbbbbbb','bbbbbbbbbbbbb','bbbb','bbbbb'],'C':['ccc'],'D':['dddd','ddd','dddddddd','ddddd']}
def f(x):
    return len(x[1])
d=dict(sorted(d.items(),key=f))
print(d)
for i,j in d.items():
    j=sorted(j,key=len)
    d[i]=j
print(d)
```

```python
s=input()
d={'i':1,'v':5,'x':10,'l':50,'c':100,'d':500,'m':1000}
ans=0
for i in range (len(s)):
    if i<(len(s)-1) and d[s[i]]<d[s[i+1]]:
        ans-=d[s[i]]
    else:
        ans+=d[s[i]]
print(ans)
```
