```python
l=eval(input("Enter: "))
for i in range (len(l)):
    temp=l.pop(i)
    sum_even=0
    sum_odd=0
    for j in range (len(l)):
        if j%2==0:
            sum_even+=l[j]
        else:
            sum_odd+=l[j]
    if sum_even==sum_odd:
        print('index=',i,'value=',temp)
    l.insert(i,temp)
```

```python
l=eval(input("Enter: "))
key=int(input("Enter key: "))
k=[]
start=0
la=len(l)//key
mod=len(l)%key
print(mod)
for i in range(key):
    if i<mod:
        k.append(l[start:start+la+1])
        start=start+la+1
    else:
        k.append(l[start:start+la])
        start=start+la
print(k)
h=[]
i=0
while i<la:
    for j in range(key):
        h+=k[j][i]
    i+=1
i=0
while i<mod:
    h+=k[i][-1]
    i+=1
print(h)
```
